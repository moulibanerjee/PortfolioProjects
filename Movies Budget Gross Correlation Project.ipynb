```python
# Read the data
df = pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
# Checking data types of the columns present
df.dtypes
```

```python
# Checking if there is any null value present in the file

for col in df.columns:
    p_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,p_missing))
    
```

```python
df = df.dropna()
```

```python
for col in df.columns:
    p_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,p_missing))
    
```

```python
# Converting the budget and gross column data type from object to integer
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Correcting the year and creating a new column taking the first 4 numbers i.e. the year of release

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df
```

```python
# sorting the gross

df.sort_values(by=['gross'],inplace = False, ascending = False )
```

```python
# Checking if the dataset has any duplicates and droping it

df.drop_duplicates()
```

```python
# Checking for any Outliers

df.boxplot(column=['gross'])
```

```python
plt.title('Budget Vs Gross')
plt.xlabel('Budget Earnings')
plt.ylabel('Gross Earnings')
plt.scatter(x=df['budget'], y=df['gross'])
plt.show()
```

```python
# plot budget vs gross using seaborn

sns.regplot(x='budget', y ='gross', data = df, scatter_kws = {"color":"red"}, line_kws = {"color":"black"})
```

```python
# Correlation
```

```python
df.corr()

# high correlation between budget and gross
```

```python
# correlation matrix

plt.title('Correlation Matrix')
plt.xlabel('Movie features')
plt.ylabel('Movie Features')
correlation_matrix = df.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
